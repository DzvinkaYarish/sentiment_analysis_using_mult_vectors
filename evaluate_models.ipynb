```python
% matplotlib inline
import io
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from polyglot.detect import Detector
from keras.preprocessing.text import text_to_word_sequence
from keras.preprocessing.sequence import pad_sequences


import pymorphy2

morph = pymorphy2.MorphAnalyzer()

plt.style.use('ggplot')
```

```python
DATA = "data/ua_reviews/"
MAX_SEQ_LENGTH = 20
```

```python
df_lst = []
for fl in os.listdir(DATA):
    df = pd.read_csv(DATA + fl)
    print(fl, len(df))
    df["keyword"] = fl[:-3]
    df_lst.append(df)
    # print(df['keyword_tone'].value_counts())
df = pd.concat(df_lst)
df.head()
```

```python
updated_df = df.drop_duplicates(['text'])
```

```python
def detect_language(text):
    try:
        return Detector(text).language.code    
    except:
        return "unknown"
```

```python
updated_df['language'] = updated_df['text'].apply(detect_language)
```

```python
print(updated_df.head())
mapper = {"neutral": 1, "negative": -1, "positive":1} 
updated_df['text'] = updated_df['text'].apply(lambda x: BeautifulSoup(str(x), "lxml").text.replace("\n", " "))
updated_df['text'] = updated_df['text'].apply(lambda x: re.sub(r'\[([^\|\]]+\|)?([^\]]+)\]',r'\2', x))
updated_df["keyword_tone"] = updated_df["keyword_tone"].apply(lambda x: mapper[x])
```

```python
def load_vec(emb_path, nmax=1000000):
    vectors = []
    word2id = {}
    with io.open(emb_path, 'r', encoding='utf-8', newline='\n', errors='ignore') as f:
        next(f)
        for i, line in enumerate(f):
            if i == 0:
                continue
            word, vect = line.rstrip().split(' ', 1)
            vect = np.fromstring(vect, sep=' ')
            assert word not in word2id, 'word found twice'
            vectors.append(vect)
            word2id[word] = len(word2id)
            if len(word2id) == nmax:
                break
    id2word = {v: k for k, v in word2id.items()}
    embeddings = np.vstack(vectors)
    return embeddings, id2word, word2id

embd, id2w, w2id = load_vec('wiki.multi.ru_lemmatized.vec')
embd.shape
```

```python
def process_data(df):
    X = [[morph.parse(w)[0].normal_form for w in text_to_word_sequence(rev)] for rev in df[df['language'] == "ru"]['text'].tolist()]
    y = np.array(df[df['language'] == "ru"]['keyword_tone'])
    return X, y
updated_df.head()
```

```python
processed_rev, y = process_data(updated_df)
X = []
for rev in processed_rev:
    emdb_rev = []
    for w in rev:
        try:
            emdb_rev.append(w2id[w])
        except KeyError:
           
            continue
    X.append(emdb_rev)
X = pad_sequences(X, maxlen=MAX_SEQ_LENGTH)
```

```python
print((X[0]))
def process_data_ua(df):
    X = [text_to_word_sequence(rev) for rev in df[df['language'] == "uk"]['text'].tolist()]
    y = None
    y = np.array(df[df['language'] == "uk"]['keyword_tone'])
    return X, y
```

```python
embd_ua, id2w_ua, w2id_ua = load_vec('data/wiki.multi.uk.vec')
processed_rev_ua, y_ua = process_data_ua(updated_df)

X_ua = []
for rev in processed_rev_ua:
    emdb_rev = []
    for w in rev:
        try:
            emdb_rev.append(w2id_ua[w])
        except KeyError:
            continue
    X_ua.append(emdb_rev)
X_ua = pad_sequences(X_ua, maxlen=MAX_SEQ_LENGTH)
X_ua.shape
```

```python
from keras.models import load_model
model1 = load_model('lstm_cnn.h5')
```

## Evaluate model for Ukrainian  language

```python
model1.evaluate(X_ua, y_ua)
```
